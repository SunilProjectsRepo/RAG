### Cosine Similarity with OpenAI Embeddings

```python
# Finding similar sentences using cosine similarity

sentences = [
    "The cat sat on the mat",
    "A feline rest on the rug",
    "The dog is in the yard",
    "I love programming in Python",
    "Python is my favorite programming language"
]

```

```python
import numpy as np

def cosine_similarity_search(vec1, vec2):
    """
    Cosine similarity measures the angle between two vectors.
    - Result close to 1: Very similar
    - Result close to 0: Not related
    - Result close to -1: Opposite meanings
    """

    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1) # get the magnitude of vector A
    norm_vec2 = np.linalg.norm(vec2) # get the magnitude of vector B

    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0

    return dot_product / (norm_vec1 * norm_vec2)
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(
    model="text-embedding-3-small"
)
sentence_embeddings = embeddings.embed_documents(sentences)
print("\nCosine Similarity Search Results:")
for i, sentence_embedding in enumerate(sentence_embeddings):
    print(f"\nSentence: '{sentences[i]}'")
```

```python
sentence_embeddings
```

```python
# Calculate the similarity between all pairs
for i in range(len(sentences)):
    for j in range(i + 1, len(sentences)):
        similarity = cosine_similarity_search(sentence_embeddings[i], sentence_embeddings[j])
        print(f"Similarity between '{sentences[i]}' and '{sentences[j]}': {similarity:.4f}")
```

```python
doc_sentences = [
    "Langchain is a powerful framework for building AI applications.",
    "RAG stands for Retrieval-Augmented Generation.",
    "Embeddings are vector representations of text.",
    "Embeddings convert text into numerical format for machine learning models.",
    "Vectors can be compared using cosine similarity to find similar texts."
]    
```

```python
# Calculate the similarity between all pairs
doc_sentence_embeddings = embeddings.embed_documents(doc_sentences)

for i in range(len(doc_sentences)):
    for j in range(i + 1, len(doc_sentences)):
        similarity = cosine_similarity_search(doc_sentence_embeddings[i], doc_sentence_embeddings[j])
        print(f"Similarity between '{doc_sentences[i]}' and '{doc_sentences[j]}': {similarity:.4f}")
```

##### Based on similarity score thresholds, we can categorize the relationships between sentences:
##### - Similarity > 0.8: Highly similar (e.g., "The cat sat on the mat" and "A feline rest on the rug")
##### - Similarity between 0.5 and 0.8: Moderately similar (e.g., "I love programming in Python" and "Python is my favorite programming language")
##### - Similarity < 0.5: Not similar (e.g., "The dog is in the yard" and "I love programming in Python")   


### Semantic Search with OpenAI Embeddings

```python
# Semantic Search - Retrieve similar documents based on a query
documents = [
    "Langchain is a framework for developing applications powered by large language models",
    "Python is a high-level programming language",
    "Machine learning is a subset of artificial intelligence",
    "Embeddings convert text into numerical vectors",
    "The weather today is sunny and warm"
]

query="What is Langchain?"
```

```python
def semantic_search(query, documents, embeddings, top_k=3):
    # embed the query and documents
    query_embedding = embeddings.embed_query(query)
    document_embeddings = embeddings.embed_documents(documents)

    # calculate the similarity score
    similarities = []
    for doc_embedding in document_embeddings:
        similarity = cosine_similarity_search(query_embedding, doc_embedding)
        similarities.append(similarity)

    print(f"Similarity scores: {similarities}")
    
    # Get top_k most similar documents
    top_indices = np.argsort(similarities)[-top_k:][::-1]
    return [(documents[i], similarities[i]) for i in top_indices]
```

```python
results = semantic_search(query, documents, embeddings)
print("\nSemantic Search Results:")
for doc, score in results:
    print(f"Document: '{doc}' with similarity score: {score:.4f}")
    
```
