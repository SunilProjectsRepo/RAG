### Hugging Face Embeddings Models

#### Sentence Transformers (Sentence to Vector Representation)

```python
from langchain_huggingface import HuggingFaceEmbeddings

# Initialize a simple Embedding model (no API key required)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

embeddings
```

```python
# Create an embedding for a sample text using embed_query method
text = "Hello, how are you? I am testing the HuggingFaceEmbeddings."
embedding_vector = embeddings.embed_query(text)
embedding_vector
print(f"Text: {text}")
print(f"Embedding vector length: {len(embedding_vector)}")
print(f"Embedding vector for the text: {embedding_vector}")
```

```python
sentences = [
    "The cat is on the table.",
    "A dog is in the garden.",
    "The sun is shining brightly.",
    "I love programming in Python.",
    "Python is a great language for data science."
]

# Create embeddings for the list of sentences using embed_documents method
sentence_embeddings = embeddings.embed_documents(sentences) 
for i, sentence in enumerate(sentences):
    print(f"Sentence: {sentence}")
    print(f"Embedding vector length: {len(sentence_embeddings[i])}")
    print(f"Embedding vector for the sentence: {sentence_embeddings[i]}\n")
```

```python
# If you want to give same sentence, you will get same embedding vector
sentences = [
    "The cat is on the table.",
    "The cat is on the table."
]
sentence_embeddings = embeddings.embed_documents(sentences)
for i, sentence in enumerate(sentences):
    print(f"Sentence: {sentence}")
    print(f"Embedding vector length: {len(sentence_embeddings[i])}")
    print(f"Embedding vector for the sentence: {sentence_embeddings[i]}\n")
```
